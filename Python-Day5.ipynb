```python
#data wrangling: numpy and pandas
#nltk for text mining
cities = {'bangalore': 50,'chennai':45,'pune': 40}
#reverse dictionary:
y = {}
#y = {50: 'bangalore',45:'chennai',40:'pune'}
for k,v in cities.items():
    y.update({v : k})
    #y[v] = k
y
```

```python
#identify common elements between two lists:
x = ['chennai','bangalore','pune']
y = ['bangalore','hyderabad','pune']
z = []
for name1 in x:
    for name2 in y:
        if name1 == name2:
            z.append(name1)
print(z)
```

```python
#other method:
for city in x:
    if city in y:
        z.append(city)
z
```

```python
#other method:
list(set(x) & set(y))
```
